```python
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
import numpy as np
import warnings
import matplotlib.lines as lines
from matplotlib.ticker import MaxNLocator
import matplotlib.gridspec as gridspec

import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True

fig = plt.figure(figsize=(5, 7))
gs = gridspec.GridSpec(2, 1, height_ratios=[4, 1])

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[1,0], sharex=ax0)
gs.update(hspace=0)

ax0.grid(True)
ax1.grid(True)

phi, II, VV, PA = np.loadtxt('dats/test/hump[1].dat', unpack = True,
            usecols = (0, 1, 2, 3))
phi = phi[::-1]
II = II[::-1]
VV = VV[::-1]
PA = PA[::-1]

IImax = max(II)
II /= IImax
VV /= IImax

# здесь исключаются все значения 'nan'
II = II[np.logical_not(np.isnan(VV))]
phi = phi[np.logical_not(np.isnan(VV))]
PA = PA[np.logical_not(np.isnan(VV))]
VV = VV[np.logical_not(np.isnan(VV))]

#ax1.set_aspect(12)
# здесь добавляется 'прозрачность' к точкам PA
#rgba_colors = []
#for i in range(len(PA)):
#    value = 1 - ((II[i])**2)
#    if value < 0:
#        value = 0
#    rgba_colors.append([value, value, value, 1.0])
#ax1.scatter(phi, PA, color = rgba_colors)

def modifyPA(PA):
    def previous(i):
        return np.average(PA[max(0, i-5):i])

    size = 90
    PA = PA[::-1]
    for i in range(1, len(PA)):
        if abs(PA[i] - previous(i)) > size:
            if PA[i] > previous(i):
                while abs(PA[i] - previous(i)) > size:
                    PA[i] -= 180.0
            else:
                while abs(PA[i] - previous(i)) > size:
                    PA[i] += 180.0
    return PA[::-1]

ax1.plot(phi, PA, c='black')
ax0.plot(phi, II, c='black')
ax0.plot(phi, VV, c='black', ls=':', lw=2)

# line = [(-5, -1), (-5, 1.5)]
# (line_xs, line_ys) = zip(*line)
# ax0.add_line(lines.Line2D(line_xs, line_ys, linewidth = 1.0, color = 'red', ls='--'))
# line = [(5, -1), (5, 1.5)]
# (line_xs, line_ys) = zip(*line)
# ax0.add_line(lines.Line2D(line_xs, line_ys, linewidth = 1.0, color = 'blue', ls='--'))
#
# ax0.text(-5, -0.2, r"$V>0$",
#             color = 'red', fontsize = 15,
#             verticalalignment='center',
#             horizontalalignment='center',
#             bbox={'facecolor':'white', 'pad':10, 'edgecolor': 'red', 'lw':0.5})
# ax0.text(5, -0.2, r"$V<0$",
#             color = 'blue', fontsize = 15,
#             verticalalignment='center',
#             horizontalalignment='center',
#             bbox={'facecolor':'white', 'pad':10, 'edgecolor': 'blue', 'lw':0.5})

# здесь подписи осей и их размеры
ax0.set_ylabel(r'$I$, $V$', fontsize=15)
ax1.set_xlabel(r'$\phi, ^\circ$', fontsize=15)
ax1.set_ylabel(r'$p.a., ^\circ$', fontsize=15)
ax0.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')
# ax1.set_yticks(range(0,180,40))
#plt.figure(figsize=(5,5))

# определение области 'y' на графике PA
#ax1.set_ylim(0.0, 180.0)

# определение области 'x' и отношения сторон на графике I, V
ax0.set(aspect = 30)
#ax0.set_ylim(-0.3, 1.1)
ax0.set_xlim(min(phi), max(phi))

# with warnings.catch_warnings():
#     warnings.simplefilter("ignore", UserWarning)
#     # This raises warnings since tight layout cannot
#     # handle gridspec automatically. We are going to
#     # do that manually so we can filter the warning.
#     gs.tight_layout(fig)

# plt.tight_layout()
# вместо 'eps' можно 'png'
plt.savefig('P02_omode_gamma1000.png', dpi=400, bbox_inches='tight')
#plt.show()

```
